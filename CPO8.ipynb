```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import scipy.signal
```

```
im = np.array(Image.open('zad5_1.JPG'))/255.0
im = np.mean(im, axis = 2)

plt.imshow(im, cmap = 'gray')
```

```
x = im

no_bins = 13
h = np.ones((1,no_bins))/no_bins

y = scipy.signal.convolve2d(x, h, 'same')
y = y + 0.01 * np.random.rand(y.shape[0], y.shape[1])

plt.figure(figsize=(12, 8))
plt.subplot(1,2,1)
plt.imshow(x, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(y, cmap = 'gray')
```

```
X = np.fft.fftshift(np.fft.fft2(x))
Y = np.fft.fftshift(np.fft.fft2(y))
H = np.fft.fftshift(np.fft.fft2(h, s = y.shape))

```

```

plt.figure(figsize= (18, 9))
plt.subplot(1,3,1)
plt.imshow(np.absolute(H))
#plt.colorbar()
plt.title('|H|')

plt.subplot(1,3,2)
plt.imshow(np.absolute(X))
plt.clim(0, 10)
#plt.colorbar()
plt.title('|X|')

plt.subplot(1,3,3)
plt.imshow(np.absolute(Y))
plt.clim(0, 10)
#plt.colorbar()
plt.title('|Y|')
```

```
plt.figure(figsize= (12, 9))
plt.subplot(2,2,1)
plt.imshow(np.absolute(H))
#plt.colorbar()
plt.title('|H|')

plt.subplot(2,2,2)
plt.imshow(np.absolute(X))
plt.clim(0, 10)
#plt.colorbar()
plt.title('|X|')

plt.subplot(2,2,3)
plt.imshow(np.absolute(Y))
plt.clim(0, 10)
#plt.colorbar()
plt.title('|Y|')

plt.subplot(2,2,4)
plt.imshow(np.absolute(H*X))
plt.clim(0, 10)
#plt.colorbar()
plt.title('|H*X|')
```

```
im = np.array(Image.open('zad5_2.JPG'))/255.0
im = np.mean(im, axis = 2)

plt.imshow(im, cmap = 'gray')
```

```
y = im

no_bins = 7
h = np.ones((no_bins, 1))/no_bins

Y = np.fft.fft2(y)
H = np.fft.fft2(h, s = y.shape)

c = 0.01
X_ = Y*np.conjugate(H)/(np.abs(H)**2+c)
x_ = np.fft.ifft2(X_)
x_ = np.real(x_)

plt.figure(figsize=(12, 8))
plt.subplot(1,2,1)
plt.imshow(y, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(x_, cmap = 'gray')
```

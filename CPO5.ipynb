# Filtry konwolucyjne

```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import scipy.signal
```

```
im = np.array(Image.open('lena.png'))/255.0
im = np.mean(im, axis = 2)

plt.figure(figsize = (6, 6))
plt.imshow(im, cmap='gray')
```

```
mask = np.ones((11,11))
mask

#Filtracja obrazu
im_f = scipy.signal.convolve2d(im, mask, 'same')

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im, cmap='gray')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_f, cmap='gray')
plt.colorbar()
```

```
print(f'Rozmiar oryginalny: {im.shape}')
print(f'Rozmiar po filtracji: {im_f.shape}')
```

#Odszumianie

```
im_n = im + np.random.normal(size = im.shape)*0.1

mask = np.ones((5,5))
im_nf = scipy.signal.convolve2d(im_n, mask, 'same')

plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
plt.imshow(im_n, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(im_nf, cmap='gray')
```

# Detekcja krawedzi

```
mask = np.array([[-1, 0, 1]])

#Filtracja obrazu
im_f = scipy.signal.convolve2d(im, mask, 'same')

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im, cmap='gray')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_f, cmap='jet')
plt.clim(-0.4, 0.4)
plt.colorbar()
```

```
mask = np.array([[-1, 0, 1]]).T

print(f'mask={mask}')

#Filtracja obrazu
im_f = scipy.signal.convolve2d(im, mask, 'same')

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im, cmap='gray')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_f, cmap='jet')
plt.clim(-0.4, 0.4)
plt.colorbar()
```

# Wyostrzanie

```
im_sharp = np.array(Image.open('lego_sharp.JPG'))/255.0
im_blur = np.array(Image.open('lego_blur.JPG'))/255.0

im_sharp = np.mean(im_sharp, axis = 2)
im_blur = np.mean(im_blur, axis = 2)

plt.figure(figsize = (12, 6))
plt.subplot(1,2,1)
plt.imshow(im_sharp, cmap='gray')
plt.clim(0, 0.7)

plt.subplot(1,2,2)
plt.imshow(im_blur, cmap='gray')
plt.clim(0, 0.7)
```

```
mask = np.array([[ 0, -1,  0],
                 [-1,  4, -1],
                 [ 0, -1,  0]])

print(f'mask={mask}')

#Filtracja obrazu
im_f = scipy.signal.convolve2d(im_blur, mask, 'same')

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im_blur, cmap='gray')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_f, cmap='jet')
plt.clim(-0.0025, 0.0025)
plt.colorbar()
```

```
mask = np.array([[ 0, -1,  0],
                 [-1,  4, -1],
                 [ 0, -1,  0]])

print(f'mask={mask}')

#Filtracja obrazu
im_f = scipy.signal.convolve2d(im_blur, mask, 'same')
im_sharp2 = im_blur + 50*im_f

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im_blur, cmap='gray')
plt.clim(0, 0.7)
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_sharp2, cmap='gray')
plt.clim(0, 0.7)
plt.colorbar()
```

```
mask_laplace = np.array([[ 0, -1,  0],
                         [-1,  4, -1],
                         [ 0, -1,  0]])*50

mask_none = np.array([[0, 0, 0],
                      [0, 1, 0],
                      [0, 0 ,0]])


#im_laplace = scipy.signal.convolve2d(im_blur, mask_laplace, 'same')
#im_none = scipy.signal.convolve2d(im_blur, mask_none, 'same')
#im_sharp2 = im_none + im_laplace

im_sharp2 = scipy.signal.convolve2d(im_blur, mask_none + mask_laplace, 'same')

plt.figure(figsize = (12, 6))

plt.subplot(1,2,1)
plt.imshow(im_blur, cmap='gray')
plt.clim(0, 0.7)
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(im_sharp2, cmap='gray')
plt.clim(0, 0.7)
plt.colorbar()
```

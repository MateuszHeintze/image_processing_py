# Wprowadzenie do cyfrowego przetwarzania obrazów
## Operacje jednopunktowe na obrazach

```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Odczytanie i wyświetlenie obrazu

```
im = np.array(Image.open('lena.jpg'))

plt.figure(figsize = (8, 8))
plt.imshow(im)
```

```
plt.figure()
plt.imshow(im)

plt.figure(figsize=(18, 6))

plt.subplot(1,3,1)
plt.imshow(im[:, :, 0], cmap='gray')
plt.clim(0, 255)
plt.colorbar()
plt.title('Red')

plt.subplot(1,3,2)
plt.imshow(im[:, :, 1], cmap='gray')
plt.clim(0, 255)
plt.colorbar()
plt.title('Green')

plt.subplot(1,3,3)
plt.imshow(im[:, :, 2], cmap='gray')
plt.clim(0, 255)
plt.colorbar()
plt.title('Blue')
```

### Konwersja do skali szarości

```
im_g = np.mean(im, axis = 2)/255.0

print('shape:', im_g.shape)
print('dtype:', im_g.dtype)

plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
```

## Operacje jednopunktowe
### Progowanie

```
im_thresh = (im_g > 0.5).astype(np.float64)

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_thresh, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Progowanie')
```

```
def threshold(im):
  im_th = (im > 0.5).astype(np.float64)
  return im_th
```

```
im_thresh2 = threshold(im_g)

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_thresh2, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Progowanie')
```

```
im_bright = im_g + 0.2

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_bright, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zwiększona jasność')
```

```
def increase_brightness(im):
  return im+0.2
```

```
im_bright2 = increase_brightness(im_g)

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_bright2, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zwiększona jasność')
```

```
im_dark = im_g - 0.2

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_dark, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zmniejszona jasność')
```

### Modyfikacja kontrastu

```
im_high = im_g * 1.4

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_high, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zwiększony kontrast')
```

```
im_low = im_g * 0.6

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_low, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zmniejszony kontrast')
```

```
def increase_contrast(im):
  return im*1.4
```

### Operacje nieliniowe

```
im_2 = im_g**2

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_2, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Korekcja gamma = 2')
```

```
def gamma_2(im):
  return im**2
```

```
im_05 = im_g**0.5

plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(im_g, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Obraz oryginalny')

plt.subplot(1,2,2)
plt.imshow(im_05, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Korekcja gamma = 0.5')
```

```
def gamma_05(im):
  return im**0.5
```

## Wizualizacja w formie wykresów

```
x = np.linspace(0, 1, 200)
y = threshold(x)

plt.plot(x, y)
plt.xlabel('Piksele oryginalne')
plt.ylabel('Piksele przetworzone')
```

```
x = np.linspace(0, 1, 200)
y = increase_brightness(x)

plt.plot(x, y)
plt.ylim(0, 1)
plt.xlabel('Piksele oryginalne')
plt.ylabel('Piksele przetworzone')
```

```
def plot_transformation(fun):
  x = np.linspace(0, 1, 200)
  y = fun(x)

  plt.plot(x, y)
  plt.ylim(0, 1)
  plt.xlabel('Piksele oryginalne')
  plt.ylabel('Piksele przetworzone')
```

```
plot_transformation(increase_brightness)
```

```
plot_transformation(increase_contrast)
```

```
plot_transformation(gamma_2)
```

```
plot_transformation(gamma_05)
```

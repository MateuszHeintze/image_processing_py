# Przetwarzanie obraz√≥w barwnych

```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import skimage.color
```

```
im = np.array(Image.open('delivery.jpg'))

plt.imshow(im)
```

```
plt.imshow(im[:, :, 0])

plt.title('Czerwony')
plt.colorbar()
plt.clim(0, 255)
```

```
plt.imshow(im[:, :, 1])

plt.title('Zielony')
plt.colorbar()
plt.clim(0, 255)
```

```
plt.imshow(im[:, :, 2])

plt.title('Niebieski')
plt.colorbar()
plt.clim(0, 255)
```

## Reprezentacja HSV

H - hue (odcie≈Ñ)
S - saturation (nasycenie)
V - value (warto≈õƒá/jasno≈õƒá)

```
im_hsv = skimage.color.rgb2hsv(im)

im_hsv.shape
#plt.imshow(im_hsv)
```

```
plt.imshow(im_hsv[:, :, 0])

plt.title('Hue (odcien)')
plt.colorbar()
plt.clim(0, 1)
```

```
plt.imshow(im_hsv[:, :, 1])

plt.title('Saturation (nasycenie)')
plt.colorbar()
plt.clim(0, 1)
```

```
plt.imshow(im_hsv[:, :, 2], cmap='gray')

plt.title('Value (jasnosc)')
plt.colorbar()
plt.clim(0, 1)
```

## Modyfikacja sk≈Çadowych HSV

```
im_hsv2 = im_hsv.copy()
im_hsv2[:,:,1] = im_hsv2[:,:,1]*1.5

im2 = skimage.color.hsv2rgb(im_hsv2)

plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.imshow(im)

plt.subplot(1,2,2)
plt.imshow(im2)
```

```
im_hsv3 = im_hsv.copy()
im_hsv3[:,:,2] = im_hsv3[:,:,2]*1.5

im3 = skimage.color.hsv2rgb(im_hsv3)

plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.imshow(im)

plt.subplot(1,2,2)
plt.imshow(im3)
```

```
im_hsv4 = im_hsv.copy()
im_hsv4[:,:,0] = im_hsv4[:,:,0]+0.1

im4 = skimage.color.hsv2rgb(im_hsv4)

plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.imshow(im)

plt.subplot(1,2,2)
plt.imshow(im4)
```

## Zmiana koloru ubrania

```
mask = np.logical_and(im_hsv[:,:,0] > 0.2, im_hsv[:,:,0] < 0.8) 

plt.imshow(mask)
plt.colorbar()
```

```
ind = np.argwhere(mask == 1)
ind.shape
```

```
rows = ind[:, 0]
cols = ind[:, 1]

im_hsv5 = im_hsv.copy()
im_hsv5[rows, cols, 0] = im_hsv5[rows, cols, 0]+0.1

im5 = skimage.color.hsv2rgb(im_hsv5)

plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
plt.imshow(im)

plt.subplot(1,2,2)
plt.imshow(im5)
```

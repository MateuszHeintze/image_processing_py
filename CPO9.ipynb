```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import scipy.signal
```

```
im = np.array(Image.open('lena.png'))/255
im = np.mean(im, axis = 2)

plt.figure(figsize=(8, 8))
plt.imshow(im)
plt.xticks(np.linspace(0,512,50))
plt.yticks(np.linspace(0,512,50))
plt.grid()
```

```
x = im
y = im[0::4, 0::4]

plt.figure(figsize=(12,12))
plt.subplot(121)
plt.imshow(x, cmap='gray')

plt.subplot(122)
plt.imshow(y, cmap='gray')
```

```
x = scipy.signal.convolve2d(im, np.ones((8,8))/64, mode='same')
y = x[0::8, 0::8]

plt.figure(figsize=(12,12))
plt.subplot(121)
plt.imshow(x, cmap='gray')

plt.subplot(122)
plt.imshow(y, cmap='gray')
```

```
plt.imshow(y, cmap='gray', interpolation='nearest')
```

```
z = np.zeros((128, 128))
z[0::2,0::2] = y
z[1::2,0::2] = y
```

```
z = np.zeros((128, 128))
z[0::2,0::2] = y
z[1::2,0::2] = y
z[0::,1::2] = z[0::,0::2]

plt.figure(figsize=(12,12))
plt.subplot(121)
plt.imshow(y, cmap='gray', interpolation='nearest')

plt.subplot(122)
plt.imshow(z, cmap='gray', interpolation='nearest')
```

```
z = np.zeros((128, 128))
z[0::2,0::2] = y
z[1:-1:2,0::2] = 0.5*z[0:-2:2,0::2] + 0.5*z[2::2,0::2]
z[0::1, 1:-1:2] = 0.5*z[0::1, 0:-2:2] + 0.5*z[0::1, 2::2]

plt.figure(figsize=(12,12))
plt.subplot(121)
plt.imshow(y, cmap='gray', interpolation='nearest')

plt.subplot(122)
plt.imshow(z, cmap='gray', interpolation='nearest')
```

```
height, width = im.shape

x = np.arange(width)
y = np.arange(height)

xx, yy = np.meshgrid(x, y)

xv = xx.ravel()
yv = yy.ravel()

fi = 15/180*np.pi

xv_ = xv*np.cos(fi) - yv*np.sin(fi)
yv_ = xv*np.sin(fi) + yv*np.cos(fi)

ii = np.round(xv_).astype(np.int32)
jj = np.round(yv_).astype(np.int32)

#np.logical_and(ii>=0, ii<width)
ind = np.argwhere((ii>=0) * (ii<width) * (jj>=0) * (jj<height))

im_rot = np.zeros((height, width))
im_rot[yv[ind], xv[ind]] = im[jj[ind], ii[ind]]

plt.figure(figsize=(8,8))
plt.subplot(121)
plt.imshow(im, cmap='gray')

plt.subplot(122)
plt.imshow(im_rot, cmap='gray')
```
